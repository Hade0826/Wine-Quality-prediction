```python
# Import Algorithms
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# Import train test split to split data
from sklearn.model_selection import train_test_split, cross_val_score

from sklearn.preprocessing import StandardScaler

# Import evaluation metrics
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score, confusion_matrix, classification_report
```

```python
#split data into training and testing datasets
x = data.drop("quality", axis=1)
y = data["quality"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
y.value_counts()
```

```python
y_dict = dict(y.value_counts())
```

```python
y_dict
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
for k, v in y_dict.items():
    per = v/len(y) *100
    print(f"class {k}, n={v}, {per:.2f}%")
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
np.random.seed(0)
#Instantiate models
sv, kn, rf = SVC(), KNeighborsClassifier(), RandomForestClassifier()

#fit my models
sv.fit(x_train, y_train)
kn.fit(x_train, y_train)
rf.fit(x_train, y_train)

#Make predictions with the models
y_pred_sv = sv.predict(x_test)
y_pred_kn = kn.predict(x_test)
y_pred_rf = rf.predict(x_test)
```

```python
print(classification_report(y_test, y_pred_sv))
```

```python
print(classification_report(y_test, y_pred_kn))
```

```python
print(classification_report(y_test, y_pred_rf))
```

### Balancing the datasets

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
```

```python
oversample = SMOTE()
undersample = RandomUnderSampler()
steps = [("o", oversample), ("u", undersample)]
pipeline = Pipeline(steps=steps)
x2, y2 = pipeline.fit_resample(x, y)
```

```python
y_dict2 = dict(y2.value_counts())
```

```python
y_dict2
```

```python
for k, v in y_dict2.items():
    per = v/len(y2) *100
    print(f"class {k}, n={v}, {per:.2f}%")
```

```python
#split data into training and testing datasets
x = data.drop("quality", axis=1)
y = data["quality"]
x_train, x_test, y_train, y_test = train_test_split(x2, y2, test_size=0.2, random_state=0)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
np.random.seed(0)
#fit my models
sv.fit(x_train, y_train)
kn.fit(x_train, y_train)
rf.fit(x_train, y_train)

#Make predictions with the models
y_pred_sv = sv.predict(x_test)
y_pred_kn = kn.predict(x_test)
y_pred_rf = rf.predict(x_test)
```

```python
print(classification_report(y_test, y_pred_sv))
```

```python
print(classification_report(y_test, y_pred_kn))
```

```python
print(classification_report(y_test, y_pred_rf))
```
